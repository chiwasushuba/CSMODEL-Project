# Dataset Description
## Family Income and Expenditure Survey 2012
 
### __Brief Description__
The 2012 Family Income and Expenditure Survey (FIES) adopted the 2009 Philippine Classification of Individual Consumption According to Purpose (PCOICOP). The 2009 PCOICOP is the first standard classification of individual consumption expenditure in the country prepared by the National Statistical Coordination Board (NSCB) in collaboration with concerned agencies in the Philippine Statistical System (PSS). The 2009 PCOICOP was patterned after the 1999 Classification of Individual Consumption According to Purpose (COICOP) issued by the United Nations Statistics Division (UNSD). The 2012 FIES is a sample survey designed to provide income and expenditure data for the country and its seventeen (17) regions. These data are the basic sources of information on human development and the basis of the policies and programs of the government to improve the socio-economic conditions of the Filipino people. 

### __Description of the Collection Process__

 The survey was conducted in two separate rounds using separate questionnaires with the same set of questions. The first was conducted in July 2012 and the second was in January 2013. For each round, the reference period was the half-year period preceding the survey. The primary strata was the 17 regions of the country and used the 2003 Master Sample (MS) as its sampling framework, which was also based on the 2000 Census of Population and Housing (CPH). With the 17 regions, within each one, another level of stratification was done using geographic subdivisions such as provinces. Highly Urbanized Cities (HUCs) and Independent Component Cities (ICCs). After this, Primary Sampling Units (PSUs) were selected by using the probability proportional to size. With this, larger PSUs had a higher chance of selection in the random sampling. After, Enumeration Areas (EAs) were chosen within each sampled PSU. Overall. the collection procedure involved a national sample of approximately 50,000 households to ensure coverage of both national and regional levels.

### Q1 - How does the average salary of the top job type in each region compare to the region's overall average salary?
* For the first question it asks the average salary of the top job type in each region compare to the region's overall average.  
* Answering how the highest paying job type in each region compares to the region's overall average salary reveals whether certain high paying jobs are driving up the regional salary averages. This insight can help people identify influential job sectors and consider the impact of supporting those sectors.
* This question revolves around the columns `W_REGN` to distinguish each regions, `OCCUP` to distinguish their jobs/occupations, and lastly `NONAGRI_SAL` to show their salaries. 

```python
# Calculate the average salary for each job type by region
avg_salary_by_job = selected_df.groupby(['W_REGN', 'OCCUP'])['NONAGRI_SAL'].mean().reset_index()
```

```python
# Get the top job type for each region based on average salary
top_job_by_region = avg_salary_by_job.loc[avg_salary_by_job.groupby('W_REGN')['NONAGRI_SAL'].idxmax()]
```

```python
# Calculate the overall average salary for each region
overall_avg_salary = selected_df.groupby('W_REGN')['NONAGRI_SAL'].mean().reset_index()
overall_avg_salary.columns = ['W_REGN', 'Overall_Avg_Salary']
```

```python
# Merge the top job salary and overall average salary
comparison_df = pd.merge(top_job_by_region, overall_avg_salary, on='W_REGN', how='inner')
```

```python
# Rename the columns for clarity
comparison_df.columns = ['W_REGN', 'Top_Occupation', 'Top_Avg_Salary', 'Overall_Avg_Salary']
```

```python
# Display the resulting DataFrame
comparison_df
```

### Numerical Sumaries

```python
# Calculate numerical summaries for Top Average Salary
top_avg_salary_mean = comparison_df['Top_Avg_Salary'].mean()
top_avg_salary_median = comparison_df['Top_Avg_Salary'].median()
top_avg_salary_std = comparison_df['Top_Avg_Salary'].std()
top_avg_salary_range = comparison_df['Top_Avg_Salary'].max() - comparison_df['Top_Avg_Salary'].min()

# Calculate numerical summaries for Overall Average Salary
overall_avg_salary_mean = comparison_df['Overall_Avg_Salary'].mean()
overall_avg_salary_median = comparison_df['Overall_Avg_Salary'].median()
overall_avg_salary_std = comparison_df['Overall_Avg_Salary'].std()
overall_avg_salary_range = comparison_df['Overall_Avg_Salary'].max() - comparison_df['Overall_Avg_Salary'].min()

# Displaying the numerical summaries
print("Numerical Summaries for Top Average Salary:")
print(f"  Mean: {top_avg_salary_mean:.2f}")
print(f"  Median: {top_avg_salary_median:.2f}")
print(f"  Standard Deviation: {top_avg_salary_std:.2f}")
print(f"  Range: {top_avg_salary_range:.2f}")

print("\nNumerical Summaries for Overall Average Salary:")
print(f"  Mean: {overall_avg_salary_mean:.2f}")
print(f"  Median: {overall_avg_salary_median:.2f}")
print(f"  Standard Deviation: {overall_avg_salary_std:.2f}")
print(f"  Range: {overall_avg_salary_range:.2f}")
```

### Binning Ages into Groups

In this step, we categorize the `AGE` variable into specific age groups to facilitate the analysis of salary distributions across different age ranges. The process involves the following steps:

1. **Define Age Bins**: 
   - We define five age bins to group the data:
     - 18-30
     - 31-40
     - 41-50
     - 51-60
     - 60+
   - These bins are set using the `age_bins` list, where the range is specified in years.

2. **Define Age Labels**: 
   - We assign labels for each bin using the `age_labels` list to represent the respective age groups.

3. **Create Age Group Column**: 
   - Using the `pd.cut()` function, we assign each individual to an age group based on their age.
   - The new column, `Age_Group`, is added to the DataFrame, where each value corresponds to one of the defined age groups.

This transformation allows us to analyze how salary distributions differ across predefined age groups, which is crucial for performing hypothesis testing, such as the Kruskal-Wallis test for non-parametric comparisons across groups.


```python
age_bins = [18, 30, 40, 50, 60, 100]  # Example bins
age_labels = ['18-30', '31-40', '41-50', '51-60', '60+']

selected_df['Age_Group'] = pd.cut(selected_df['AGE'], bins=age_bins, labels=age_labels, right=False)
selected_df.head()
```

### Kruskal-Wallis Test for Salary and Age Groups

In the analysis, we used the non-parametric Kruskal-Wallis Test to check whether the significant existence of differences in the salaries is in the different groups of age. The Kruskal-Wallis Test doesn't necessarily need a normal distribution of data since often data on salaries are skewed and will thus usually contain outliers. This is an important requirement for our study. Since the salary data at different age groups may not be normally distributed, this test offers a powerful approach to compare the salary distributions across age groups.

The Kruskal-Wallis Test will determine whether there are significant differences in salary distributions among the predefined age groups. This addresses the research question by analyzing how the differences in salaries between the age groups are statistically significant or not. That is, whether the differences are large enough to reject the null hypothesis. In a nutshell, the Kruskal-Wallis Test allows us to make meaningful inferences about how age relates to the salary distributions, providing statistical evidence behind decisions that will be made based on age-based trends in the salary data.

#### Hypotheses for Age and Salary

##### Salary Distributions by Age Group
***$H_0$:*** There is no significant difference in salary distributions across different age groups.

***$H_A$:*** There is a significant difference in salary distributions across different age groups.

The hypotheses for salary distributions focus on whether the salary distributions differ significantly across the various age groups.

#### Hypothesis Testing

After completing the necessary data preprocessing steps, the dataset is now prepared for hypothesis testing. This process will help to determine if there are significant differences in salary distributions across different age groups, as hypothesized.

```python
from scipy.stats import kruskal

salary_by_age_group = [selected_df[selected_df['Age_Group'] == age_group]['NONAGRI_SAL'] for age_group in age_labels]

stat, p_value = kruskal(*salary_by_age_group)

print(f"Kruskal-Wallis H-statistic: {stat}")
print(f"P-value: {p_value}")

alpha = 0.05 
if p_value < alpha:
    print("We reject the null hypothesis. There are significant differences in salary distributions across age groups.")
else:
    print("We fail to reject the null hypothesis. There are no significant differences in salary distributions across age groups.")
```

The Kruskal-Wallis test of the distributions of the salaries by age groups results in a statistically significant difference found between the groups (H = 152.50, p < 0.001). Indeed, this tells us that there are significant differences among the salary distributions across the different age groups; therefore, we reject the null hypothesis of no significant difference in salary distributions among the groups.
