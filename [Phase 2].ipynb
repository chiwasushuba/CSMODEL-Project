# Dataset Description
## Family Income and Expenditure Survey 2012
 
### __Brief Description__
The 2012 Family Income and Expenditure Survey (FIES) adopted the 2009 Philippine Classification of Individual Consumption According to Purpose (PCOICOP). The 2009 PCOICOP is the first standard classification of individual consumption expenditure in the country prepared by the National Statistical Coordination Board (NSCB) in collaboration with concerned agencies in the Philippine Statistical System (PSS). The 2009 PCOICOP was patterned after the 1999 Classification of Individual Consumption According to Purpose (COICOP) issued by the United Nations Statistics Division (UNSD). The 2012 FIES is a sample survey designed to provide income and expenditure data for the country and its seventeen (17) regions. These data are the basic sources of information on human development and the basis of the policies and programs of the government to improve the socio-economic conditions of the Filipino people. 

### __Description of the Collection Process__

 The survey was conducted in two separate rounds using separate questionnaires with the same set of questions. The first was conducted in July 2012 and the second was in January 2013. For each round, the reference period was the half-year period preceding the survey. The primary strata was the 17 regions of the country and used the 2003 Master Sample (MS) as its sampling framework, which was also based on the 2000 Census of Population and Housing (CPH). With the 17 regions, within each one, another level of stratification was done using geographic subdivisions such as provinces. Highly Urbanized Cities (HUCs) and Independent Component Cities (ICCs). After this, Primary Sampling Units (PSUs) were selected by using the probability proportional to size. With this, larger PSUs had a higher chance of selection in the random sampling. After, Enumeration Areas (EAs) were chosen within each sampled PSU. Overall. the collection procedure involved a national sample of approximately 50,000 households to ensure coverage of both national and regional levels.

### __Implications of the Collection Process__

As two stage stratification was used, this ensured that all regional characteristics can be captured by the survey which helps the reliability and accuracy of regional comparisons among these. As mentioned, Primary Sampling Units (PSUs) were selected by using the probability proportional to size, though this helped in efficiently collecting the data, it may affect the comparison of regions in terms of smaller PSUs or less populous areas to the more populous areas which can be called sampling bias in this situation.

### Q1 - How does the average salary of the top job type in each region compare to the region's overall average salary?
* For the first question it asks the average salary of the top job type in each region compare to the region's overall average.  
* Answering how the highest paying job type in each region compares to the region's overall average salary reveals whether certain high paying jobs are driving up the regional salary averages. This insight can help people identify influential job sectors and consider the impact of supporting those sectors.
* This question revolves around the columns `REGION` to distinguish each regions, `OCCUP` to distinguish their jobs/occupations, and lastly `NONAGRI_SAL` to show their salaries. 

```python
# Calculate the average salary for each job type by region
avg_salary_by_job = selected_df.groupby(['REGION', 'OCCUP'])['NONAGRI_SAL'].mean().reset_index()
```

```python
# Get the top job type for each region based on average salary
top_job_by_region = avg_salary_by_job.loc[avg_salary_by_job.groupby('REGION')['NONAGRI_SAL'].idxmax()]
```

```python
# Calculate the overall average salary for each region
overall_avg_salary = selected_df.groupby('REGION')['NONAGRI_SAL'].mean().reset_index()
overall_avg_salary.columns = ['REGION', 'Overall_Avg_Salary']
```

```python
# Merge the top job salary and overall average salary
comparison_df = pd.merge(top_job_by_region, overall_avg_salary, on='REGION', how='inner')
```

```python
# Rename the columns for clarity
comparison_df.columns = ['REGION', 'Top_Occupation', 'Top_Avg_Salary', 'Overall_Avg_Salary']
```

```python
# Display the resulting DataFrame
comparison_df
```

### Numerical Sumaries

```python
# Calculate numerical summaries for Top Average Salary
top_avg_salary_mean = comparison_df['Top_Avg_Salary'].mean()
top_avg_salary_median = comparison_df['Top_Avg_Salary'].median()
top_avg_salary_std = comparison_df['Top_Avg_Salary'].std()
top_avg_salary_range = comparison_df['Top_Avg_Salary'].max() - comparison_df['Top_Avg_Salary'].min()

# Calculate numerical summaries for Overall Average Salary
overall_avg_salary_mean = comparison_df['Overall_Avg_Salary'].mean()
overall_avg_salary_median = comparison_df['Overall_Avg_Salary'].median()
overall_avg_salary_std = comparison_df['Overall_Avg_Salary'].std()
overall_avg_salary_range = comparison_df['Overall_Avg_Salary'].max() - comparison_df['Overall_Avg_Salary'].min()

# Displaying the numerical summaries
print("Numerical Summaries for Top Average Salary:")
print(f"  Mean: {top_avg_salary_mean:.2f}")
print(f"  Median: {top_avg_salary_median:.2f}")
print(f"  Standard Deviation: {top_avg_salary_std:.2f}")
print(f"  Range: {top_avg_salary_range:.2f}")

print("\nNumerical Summaries for Overall Average Salary:")
print(f"  Mean: {overall_avg_salary_mean:.2f}")
print(f"  Median: {overall_avg_salary_median:.2f}")
print(f"  Standard Deviation: {overall_avg_salary_std:.2f}")
print(f"  Range: {overall_avg_salary_range:.2f}")
```

We need to scale these data because the features with larger ranges such as salary can dominate the model based on result and this can affect the data modelling. Scaling normalizes the features ensuring that each one can contribute equally to the model.

This performs an Elbow Method to help determine the optimal number of clusters for the k-means clustering algorithm. The Elbow method is a technique used to determine the optimal number of clusters for clustering algorithms like the KMeans Algorithm. It helps identify the point at which the increasing number of clusters provide better result.

* The `k_values` ranging 1 to 10 is the default k_values when performing the elbow method to determine the most optimal k or number of clusters.
* for the `random_state` we used 0 as it's value so that the output can be reproducable again instead of having different value every run.

```python
inertia_values = []
k_values = range(1, 11)  

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(train_df)
    inertia_values.append(kmeans.inertia_)
```

#### Visualization:

```python
plt.figure(figsize=(10, 6))
plt.plot(k_values, inertia_values, marker='o', linestyle='-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

Based on the graph of the elbow method we can see that the optimal `n` for our data is `3` because of the sudden change in direction.

K-Means Algorithm
* K-Means Clustering is ideal for creating clusters of individuals based on attributes like region, occupation, salary, and other features in your data. The goal of your research is to identify distinct clusters of sole earners and understand how their salary distributions vary based on these attributes. K-Means can help you segment the data into meaningful clusters based on similar characteristics.

* `n_clusters` is the number of clusters that we want in our data it will partition the data in distinct groups. The most optimal number of clusters for the data is `3`. It's important to determine how many cluster is essential as it might lead to oversimplified groupings or overfitting. We used the elbow method to determine the optimal number of clusters.
* `init` controls the initialization method for the centroids. `"k-means++"` is the default and recommended for faster convergence. It initializes centroids by selecting points that are spread out. This reduces the number of iterations that is needed.
* `n_init` this specifies the number of times the algorithm will be run with different initialization of centroids. Running it multiple times helps in finding a better solution as K-means clustering can converge to different results depending on the initial centroids. settings the n_init to `5` will make it run 5 times each with different initial centroids.

```python
kmeans = KMeans(n_clusters=3, init='k-means++', n_init=5)
train_df
```

The followings steps consecutively happens when `.fit()` is used: Initialize Centroids, Assign points to clusters, Update centroids, Iterate until convergence. ### Shapiro-Wilk Test for Salary distribution by Region

We can further verify the results for the region's data by using the **Shapiro-Wilk Test**, which calculates the p-value to determine whether each group of data is normally distributed. The null hypothesis of the Shapiro-Wilk test asserts that the data follows a normal distribution, while the alternative hypothesis suggests that the data does not follow a Gaussian distribution. To perform this, we can utilize the **shapiro** function from the scipy library.

```python
def shapiro_test(group):
    stat, p_value = stats.shapiro(group['NONAGRI_SAL'])
    return pd.Series({'Shapiro-Wilk Statistic': stat, 'p-value': p_value})

shapiro_results = region_selected_df.groupby('REGION').apply(shapiro_test, include_groups=False)

print(shapiro_results)
```

Since the p-values for the salary distribution across regions are all less than the alpha level of 0.05 (e.g., 2.580539e-18 < 0.05 for ARMM, 5.935963e-33 < 0.05 for Bicol Region, etc.), this indicates that the salary distribution across all 16 regions is **not normal**.

# Hypotheses

### Hypothesis for Region

***$H_0$:*** *There is no significant difference in the mean salary of sole earners among the different regions.*

***$H_A$:*** *There is significant difference in the mean salary of sole earners among the different regions.*

### Hypothesis for Age
***$H_0$:*** *There is no significant difference in salary distributions across different age values.*

***$H_A$:*** *There is a significant difference in salary distributions across different age values.*

## Hypothesis Testing 

After performing the necessary data preprocessing steps, the dataset is now ready for hypothesis testing to determine whether there are significant differences in salary distributions across various factors such as age, and region, as hypothesized to supplement the data in answering our research question

### Choosing the Significance Level
Before conducting hypothesis testing, we need to define the significance level (α), which determines the threshold for rejecting the null hypothesis. In this case, we choose a significance level of 0.05. This means that we are willing to accept a 5% chance of rejecting the null hypothesis when it is actually true (Type I error). If the p-value obtained from the test is less than or equal to 0.05, we reject the null hypothesis and conclude that there are significant differences in salary distributions across the factors being tested.

### Kruskal-Wallis Test
In this study, we make use of the non-parametric Kruskal-Wallis Test to identify whether the salaries of different groups classified by age and region differ significantly. The present study is particularly apt for Kruskal-Wallis Test since its non-parametric nature does not rely on a normal distribution of data, which in most cases is an important consideration since salaries would show skewness and include outliers.

The Kruskal-Wallis Test will assess whether there are significant differences in salary distributions among the predefined groups for age and region. This addresses the research question by analyzing how salary distributions vary across these two variables. By determining whether these differences are statistically significant, we can assess whether the observed variations are large enough to reject the null hypothesis.

##### Salary Distribution Across Regions
We will now be conducting the Kruskal-Wallis test in order to identify the differences in salary distribution among the various regions. This test will help determine whether there are significant differences in salary across the regions, given that the data may not follow a normal distribution.

```python
R0 = data_modelling_df[data_modelling_df['Cluster'] == 0]['REGION']
R1 = data_modelling_df[data_modelling_df['Cluster'] == 1]['REGION']
R2 = data_modelling_df[data_modelling_df['Cluster'] == 2]['REGION']

result = stats.kruskal(R0, R1, R2)
result
```

The Kruskal-Wallis test on the distributions of salary data across **different regions** reveals a statistically insignificant difference (H = 0.8676, p = 0.6481). This suggests that salary distributions do not differ significantly between the regions, indicating that geographical location does not have a significant impact on salaries. As a result, we fail to reject the null hypothesis, which states that there is no significant difference in salary distributions among the regions, and conclude that the region does not play a significant role in the variation of salary distributions.

##### Salary Distribution Across Ages
We will now be conducting the Kruskal-Wallis test to identify the differences in salary distribution across various age groups. This test will help determine whether there are significant variations in salary between different age categories, considering that the data may not follow a normal distribution.

```python
D0 = data_modelling_df[data_modelling_df['Cluster'] == 0]['AGE']
D1 = data_modelling_df[data_modelling_df['Cluster'] == 1]['AGE']
D2 = data_modelling_df[data_modelling_df['Cluster'] == 2]['AGE']

result = stats.kruskal(D0, D1, D2)
result
```

The Kruskal-Wallis test on the distributions of **ages** across the clusters reveals a statistically significant difference (H = 6887.33, p = 0.0). This result indicates that there are significant differences in age distributions between the clusters. Therefore, we reject the null hypothesis of no significant difference in age distributions among the clusters.

These findings suggest that while factors like age and region influence salary distributions, they do not adhere to a normal distribution and thus require non-parametric methods for analysis.

Based on the results and overall analysis, this research highlights that salary distributions can differ based on region and age range. However, the Kruskal-Wallis test indicates that there is no statistically significant difference in salary distributions across regions (p > 0.05), suggesting that factors other than geographical location may play a role in salary variation. In contrast, different age groups do show significant differences in salary distributions, with younger and newer earners generally receiving lower salaries, likely due to limited experience. Those in their prime working years (ages 41-50) tend to earn higher salaries, likely due to greater experience.
